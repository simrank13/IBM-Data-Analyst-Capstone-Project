### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
## Write your code here
hist_data =  df['ConvertedCompYearly']

plt.figure(figsize=(10,8))
plt.hist(hist_data, bins=20)
plt.title('Distribution of yearly compensation')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('Distribution Frequency')
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
## Write your code here
age_map = {
    'Under 18 years old':17,
    '18-24 years old':21,
    '25-34 years old':30,
    '35-44 years old':40,
    '45-54 years old':50,
    '55-64 years old':60,
    '65 years or older':70
}

df['Age_Num'] = df['Age'].map(age_map)
box_plt_data = df['Age_Num'].dropna()


plt.figure(figsize=(12,6))
plt.boxplot(box_plt_data)
plt.title('Box Plot of Age')
plt.xlabel('Distribution')
plt.ylabel('Age')
plt.show()
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
## Write your code here
df.columns.to_list()
scatter_data = df['Age_Num'], df['ConvertedCompYearly']

plt.figure(figsize=(10,8))
plt.scatter(scatter_data[0], scatter_data[1])
plt.title('Scatter Plot of Age and ConvertedCompYearly')
plt.xlabel('Age')
plt.ylabel('ConvertedCompYearly')
plt.grid(True)
plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
## Write your code here
bubble_data =  df['ConvertedCompYearly'], df['JobSatPoints_6'], df['Age_Num']

plt.figure(figsize=(10,8))
plt.scatter(bubble_data[0], bubble_data[1], s=bubble_data[2]*10, edgecolor='yellow')
plt.title('Bubble Plot of ConvertedCompYearly and JobSatPoints_6 with Age as Bubble Size')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('JobSatPoints_6')
plt.grid(True)
plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
## Write your code here
horizontal_bar_data = df['MainBranch'].value_counts()

plt.figure(figsize=(10,8))
plt.barh(horizontal_bar_data.index, horizontal_bar_data.values)
plt.title('Horizontal Bar Chart of MainBranch Distribution')
plt.xlabel('Distribution of Respondents')
plt.ylabel('MainBranch')
plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
## Write your code here
vertical_bar_data = df['LanguageWantToWorkWith'].str.split(';').explode().value_counts().head()

plt.figure(figsize=(10,6))
vertical_bar_data.plot(kind='bar')
plt.title('Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With')
plt.xlabel('Programming Languages')
plt.ylabel('Number of Respondents')
plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
## Write your code here
stacked_bar_data = df.groupby('Age_Num')[['JobSatPoints_6', 'JobSatPoints_7']].median()

stacked_bar_data.plot(kind='bar', stacked=True)
plt.title('Stacked Bar Chart of Median JobSatPoints_6 and JobSatPoints_7 by Age Group')
plt.xlabel('Age Group')
plt.xticks(rotation = 0)
plt.ylabel('Median Job Satisfaction')
plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
## Write your code here
bar_chart_data = df['DatabaseHaveWorkedWith'].str.split(';').explode().value_counts().head(20)

plt.figure(figsize=(10,6))
bar_chart_data.plot(kind='bar')
plt.title('Bar Chart of Database Popularity')
plt.xlabel('Database')
plt.ylabel('Number of Respondents')
plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
## Write your code here
bar_chart_median_data = df.groupby('Age_Num')['ConvertedCompYearly'].median()

plt.figure(figsize=(10,6))
bar_chart_median_data.plot(kind='bar')
plt.title('Grouped Bar Chart of Median ConvertedCompYearly for Different Age Groups')
plt.xlabel('Age Group')
plt.xticks(rotation = 0)
plt.ylabel('Median ConvertedCompYearly')
plt.show()
```

##### 2. Bar Chart of Respondent Count by Country


Show the distribution of respondents by country to see which regions are most represented.



```python
## Write your code here
bar_chart_by_country_data = df['Country'].value_counts().head(20)

plt.figure(figsize=(10,8))
bar_chart_by_country_data.plot(kind='bar')
plt.title('Bar Chart of Respondent Count by Country')
plt.xlabel('Country')
plt.xticks(rotation = 45, ha = 'right')
plt.ylabel('Distribution of Respondents')
plt.tight_layout()
plt.show()
```
