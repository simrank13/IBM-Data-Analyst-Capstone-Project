### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
##Write your code here
stacked_chart1 = df.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median().plot(kind='bar', stacked=True)

plt.title('Median Job Satisfaction Scores by Age Group')
plt.xlabel('Age')
plt.ylabel('Median Score')
plt.show()
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
##Write your code here
top_20_employment = df['Employment'].value_counts().head(20).index.tolist()
print(top_20_employment )

excluded = ['Retired', 'Student, full-time',
            'I prefer not to say',
            'Not employed, but looking for work',
            'Not employed, and not looking for work',
            'Student, part-time',
            'Student, full-time;Not employed, and not looking for work',
            'Student, full-time;Not employed, but looking for work'
            ]
top_20_employment = [x for x in top_20_employment if x not in excluded]
df_top_employment = df[df['Employment'].isin(top_20_employment)]

stacked_chart2 = df_top_employment.groupby('Employment')[['JobSatPoints_6', 'JobSatPoints_7']].median().plot(kind='bar', stacked=True)

plt.title('Median Job Satisfaction Scores by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Median Score')
plt.show()
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
##Write your code here
age_map = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 32,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70,
    'Prefer not to say': 0
}
df['AgeNum'] = df['Age'].map(age_map)

age_filtered_df = df[(df['AgeNum'] >= 30) & (df['AgeNum'] <= 35)]
age_filtered_df.head()

import numpy as np

age_filtered_df = age_filtered_df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6'])

compensation = [0, 50000, 100000, 150000, 200000, np.inf]
labels = ['0-50k', '50k-100k', '100k-150k', '150k-200k', '200k+']

age_filtered_df['CompensationRange'] = pd.cut(age_filtered_df['ConvertedCompYearly'], bins=compensation, labels=labels)

age_filtered_df.loc[:, ['ConvertedCompYearly', 'JobSatPoints_6']] = age_filtered_df[['ConvertedCompYearly', 'JobSatPoints_6']]

stacked_chart3 = age_filtered_df.groupby('CompensationRange')[['ConvertedCompYearly', 'JobSatPoints_6']].median().plot(kind='bar', stacked=True)

plt.title('Compensation and Job Satisfaction by Age Group between 30 - 35')
plt.xlabel('Compensation Range')
plt.ylabel('Median Value')
plt.show()
```

##### Stacked Chart of Median Compensation and Job Satisfaction Across Age Group


Compare the median compensation and job satisfaction metrics across different age groups. This helps visualize how compensation and satisfaction levels differ by age.


```python
##Write your code here
df['Age'] = pd.Categorical(df['Age'], categories=['Under 18 years old', '18-24 years old', '25-34 years old', '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older', 'Prefer not to say'], ordered=True)

median_values_by_age = df.groupby('Age')[['ConvertedCompYearly', 'JobSatPoints_6']].median()

stacked_chart4 = median_values_by_age.plot(kind='bar', stacked=True)

plt.title('Median Compensation and Job Satisfaction by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
##Write your code here
df_db_wish_to_learn = df.groupby('AgeNum')['DatabaseWantToWorkWith'].value_counts().unstack().fillna(0)

top5_databases = df['DatabaseWantToWorkWith'].value_counts().head().index.tolist()

df_db_wish_to_learn = df_db_wish_to_learn[top5_databases]

stacked_chart5 = df_db_wish_to_learn.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Top 5 Preferred Databases to Learn by based on different Ages')
plt.xlabel('Ages')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=0)
plt.legend(title='Types of Database')
plt.tight_layout()
plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
##Write your code here
employment_job_sat = df.groupby('JobSatPoints_7')['Employment'].value_counts().unstack().fillna(0)

top_employment_types = df['Employment'].value_counts().nlargest(10).index.tolist()

valid_top_employment_types = [emp_type for emp_type in top_employment_types if emp_type in employment_job_sat.columns]

employment_job_sat_filtered = employment_job_sat[valid_top_employment_types]

stacked_chart6 = employment_job_sat_filtered.plot(kind='bar', stacked=True, figsize=(15, 10))

plt.title('Distribution of Employment Types by Job Satisfaction Level')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Number of Respondents')
plt.legend(title='Employment Type')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
##Write your code here
df_lang_admired = df.groupby('AgeNum')['LanguageAdmired'].value_counts().unstack().fillna(0)

top5_languages = df['LanguageAdmired'].value_counts().head().index.tolist()

df_lang_admired = df_lang_admired[top5_languages]

stacked_chart7 = df_lang_admired.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Top 5 Preferred Programming Languages by based on different Ages')
plt.xlabel('Ages')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=0)
plt.legend(title='Types of Programming Languages')
plt.show()
```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python
##Write your code here
platform_admired_employment = df.groupby('Employment')['PlatformAdmired'].value_counts().unstack().fillna(0)

top_10_employment_types = df['Employment'].value_counts().nlargest(10).index.tolist()

valid_top_employment_types = [emp_type for emp_type in top_10_employment_types if emp_type in platform_admired_employment.index]
platform_admired_employment_filtered_employment = platform_admired_employment.loc[valid_top_employment_types]

top_10_platforms_admired = df['PlatformAdmired'].value_counts().nlargest(10).index.tolist()

valid_top_platforms_admired = [platform for platform in top_10_platforms_admired if platform in platform_admired_employment_filtered_employment.columns]
platform_admired_employment_filtered_platforms = platform_admired_employment_filtered_employment[valid_top_platforms_admired]


stacked_chart8 = platform_admired_employment_filtered_platforms.plot(kind='bar', stacked=True, figsize=(15, 10))

plt.title('Distribution of Top 10 Admired Platforms for Top 10 Employment Types')
plt.xlabel('Employment Type')
plt.ylabel('Number of Respondents')
plt.legend(title='Platform Admired', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
